# Group and Visualize the Data

```python
length_by_area = df.groupby('Shape_Area')['Shape_Length'].sum() #aggregate the number of acres by the case's modified date
length_by_area.head(10)
```

This groups the shape areas by the shape lengths. Basically, in these mining claims, the area of the shape is a lot smaller than the length of the shape.  

```python
import matplotlib.pyplot as plt
```

```python
length_by_area.plot()
plt.title("Open Mining Claims, Plot Length by Area")
plt.xlabel("Plot Area")
plt.ylabel("Plot Length")
```

Basically, this shows that there are a lot of unsettled claims where the mining area is between 0.000 and 0.002, but the length of these claims is extremeely variable, between 0 and 7. Conversely, an area around 0.010 is in higher demand, but has a shorter length to mine, between 1 and 5. This shows that a lot of the claims are made for narrow, long areas to mine. 

```python
status_counts = df['CSE_DISP'].value_counts()
```

```python
plt.figure(figsize=(8,5))
status_counts.plot(kind='bar')

plt.title("Mining Claims by Status")
plt.xlabel("Status")
plt.ylabel("Number of Claims")
plt.xticks(rotation=45)  # rotate labels if needed
plt.tight_layout()
plt.show()
```

This highlights that the government hasn't handled any of the active cases. There are over 400,000 active claims, around 150,000 that were filed, and a miniscule amount that are under review. The submitted tally is also very tiny. This bar chart is showing that a lot of the cases are still active and waiting on further action.
